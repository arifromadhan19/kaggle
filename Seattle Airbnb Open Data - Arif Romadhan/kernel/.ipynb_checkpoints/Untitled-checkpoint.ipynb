```python
original = df_listings['calendar_updated'].drop_duplicates()
original = original.values

regex_pattern = '(\d+)(?:\s)([a-z\s]+)'

modified = []
for i in range(len(original)):
    text = original[i]
    if text == 'yesterday':
        modified.append(2)
    elif text == 'today':
        modified.append(1)
    elif text == 'a week ago':
        modified.append(7)
    elif text == 'never':
        modified.append(0)
    else:
        match = re.match(regex_pattern, text)
        if match.group(2)[0:3] == 'day':
            modified.append (int(match.group(1)))
        elif match.group(2)[0:4] == 'week':
            modified.append (int(match.group(1))*7)
        elif match.group(2)[0:5] == 'month':
            modified.append (int(match.group(1))*30)

for i in range(len(modified)):
    print ('{} = {}'.format(original[i], modified[i]))
    
    
# def convert(df):
    
```
